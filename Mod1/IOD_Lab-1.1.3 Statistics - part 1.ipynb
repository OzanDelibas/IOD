# Statistics - Part 1

```python
import numpy as np
import matplotlib.pyplot as plt
# get matplot to show charts inline
%matplotlib inline
```

```python
# frequency of some observations using pie chart
labels = 'Developers', 'Data Engineers', 'Data Scientists', 'Business Analysts'
frequency = [15, 32, 12, 7]
fig1, ax1 = plt.subplots()
explode = (0, 0, 0.3, 0)  # modify this to "explode" only the 3rd slice (i.e. 'Data Scientists')
ax1.pie(frequency, explode=explode, labels=labels, autopct='%1.2f',
        shadow=True, startangle=180)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# plot frequency of some observations using bar chart
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]
 
# Create a bar plot of performance for each object
plt.xticks(y_pos, objects)
plt.bar(y_pos, height=performance)

# Use ylabel and title to label the y axis and provide a title
plt.ylabel('performance')
plt.title('observations')
```
